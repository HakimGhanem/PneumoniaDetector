```python
import numpy as np
import pandas as pd

from tensorflow.keras.applications.inception_v3 import InceptionV3, preprocess_input, decode_predictions
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, BatchNormalization
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing import image

# Data preparation
# training data

batch_size = 160

train_data = image.ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    width_shift_range=0.1,
    height_shift_range=0.1
)

train_gen =train_data.flow_from_directory(
        train_path, #directory that contains training data
        target_size=(150, 150), #image size we are importing
        batch_size=batch_size, #how many files to read in at a time
        class_mode="categorical")

```

```python
# testing data

test_data = image.ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    width_shift_range=0.1,
    height_shift_range=0.1
)

test_gen = test_data.flow_from_directory(
        test_path,
        target_size=(150, 150),
        batch_size=10,
        class_mode="categorical",
        shuffle=False)


# validating data

val_data = image.ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    width_shift_range=0.1,
    height_shift_range=0.1
)

val_gen = val_data.flow_from_directory(
        val_path,
        target_size=(150, 150),
        batch_size=1,
        class_mode="categorical",
        shuffle=False)
```

```python
trained_model.compile(loss ="categorical_crossentropy", 
                    optimizer = "adam", 
                    metrics=["accuracy"])
```

```python
#Fiiting and Testing Model
fit_model = trained_model.fit_generator(
        train_gen, #train data generator 
        steps_per_epoch=len(train_gen),
        epochs=5,
        validation_data=test_gen, #validation data generator 
        validation_steps=len(test_gen))
```

```python
result  = trained_model.evaluate_generator(val_gen,steps=len(val_gen))
print('Loss:', result[0])
print('Accuracy: {:.4}%'.format(result[1]*100))
Loss: 1.4120792818503105
Accuracy: 68.75%

# Generating all results and comparing them in a data frame

files = val_gen.filenames
nb_samples = len(files)
predict = trained_model.predict_generator(val_gen,steps = nb_samples).argmax(axis=-1)
```
